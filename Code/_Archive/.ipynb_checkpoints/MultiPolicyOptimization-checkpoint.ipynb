**Functions to setup and optimize multi-policy static weighted SAA with linear decision rules**

```python
### Function to set up the model
def setUpModel(I0, xi_hat, w_hat, model_params):
    
    ### Model coefficients
    
    # dim(A) = T x T x M; access via A[t,s,m]
    A = np.array([np.array([(np.array([model_params['h'],-model_params['b']])
                             if s==t
                             else np.array([model_params['h'],-model_params['b']]))
                            if s<=t
                            else np.array([0,0])
                            for s in range(model_params['T'])])
                  for t in range(model_params['T'])])

    # dim(B) = T x T x M; access via B[t,s,m]
    B = np.array([np.array([np.array([-model_params['h'],model_params['b']])
                            if s<=t
                            else np.array([0,0])
                            for s in range(model_params['T'])])
                  for t in range(model_params['T'])])

    # dim(C) = T x T x M; access via C[t,s,m]
    C = np.array([np.array([np.array([-1,-1])
                            if s==t
                            else np.array([0,0])
                            for s in range(model_params['T'])])
                  for t in range(model_params['T'])])

    # dim(f) = T x M; access via f[t,m]
    f = np.array([np.array([-model_params['h']*I0,model_params['b']*I0])
                  for t in range(model_params['T'])])
       
  
    ### Create new model
    m = gp.Model()
    m.setParam('LogToConsole', model_params['LogToConsole'])
    m.setParam('NonConvex', model_params['NonConvex'])
    m.setParam('PSDTol', model_params['PSDTol'])
    m.setParam('MIPGap', model_params['MIPGap']) 
    

    ## Primary decision variable (q)
    q0 = m.addVars(model_params['T'], lb=0, vtype='I', name='q0') # T vector
    Q = m.addVars(model_params['T'], model_params['T'], lb=0, vtype='C', name='Q') # TxT matrix

    ## Auxiliary decision variable (y x i)
    yi0 = m.addVars(model_params['N'], model_params['T'], vtype='I', name='yi0') # NxT vector
    Yi = m.addVars(model_params['N'], model_params['T'], model_params['T'], vtype='C', name='Yi') # NxTxT matrix

    ## Fixed cost of ordering
    ki_flag = m.addVars(model_params['N'], model_params['T'], vtype='B') # NxT vector
           
            
    ### Add constraints -- non-anticipative decision rule matrices

    # Q
    if model_params['decision_rules']=='LDR':
        
        # Set matrix entries of Q to zero where s>=t to ensure non-anticpiative decisions
        Q_cons = m.addConstrs((Q[t,s] == 0)
                              for s in range(model_params['T'])
                              for t in range(model_params['T'])
                              if (s >= t))
    else:
        
        # Set all matrix entries of Q to zero
        Q_cons = m.addConstrs((Q[t,s] == 0)
                              for s in range(model_params['T'])
                              for t in range(model_params['T']))

    # Yi
    if model_params['decision_rules']=='LDR':
        
        # Set matrix entries of Y_i to zero where s>t to ensure non-anticpiative decisions
        Yi_cons = m.addConstrs((Yi[i,t,s] == 0)
                               for s in range(model_params['T']) 
                               for t in range(model_params['T']) 
                               for i in range(model_params['N'])
                               if (s > t))
    else:
        
        # Set matrix entries of Y_i to zero
        Yi_cons = m.addConstrs((Yi[i,t,s] == 0)
                               for s in range(model_params['T']) 
                               for t in range(model_params['T']) 
                               for i in range(model_params['N']))
   
    ### Add contraints -- non-negative decision space for primary decision rule
    q_cons = m.addConstrs(q0[t]+gp.quicksum(Q[t,s]*xi_hat[i,s]
                                            for s in range(model_params['T'])) >= 0
                          for t in range(model_params['T'])
                          for i in range(model_params['N']))

    ### Add constraints -- model constraints
    cons = m.addConstrs(

        # A[t,.,m]*(q0[t]+Q[.,ts]*xi[i,ts]) where A[t,s,m]==0 for s > t
        gp.quicksum(A[t,s,m]*(q0[s]+gp.quicksum(Q[s,ts]*xi_hat[i,ts]
                                                for ts in range(model_params['T'])))
                    for s in range(model_params['T'])) + 

        # B[t,.,m]*xi[i,.] where B[t,s,m]==0 for s > t
        gp.quicksum(B[t,s,m]*xi_hat[i,s]
                    for s in range(model_params['T'])) + 

        # C[t,.,m]*(y0i[t]+Yi[.,ts]*xi[i,ts]) where C[t,s,m]==0 for s != t
        gp.quicksum(C[t,s,m]*(yi0[i,s]+gp.quicksum(Yi[i,s,ts]*xi_hat[i,ts]
                                                   for ts in range(model_params['T'])))
                    for s in range(model_params['T']))

        
        <= f[t,m]

        for t in range(model_params['T'])
        for m in range(model_params['M'])
        for i in range(model_params['N'])
    )
    
    # Fixed ordering cost
    ki_cons1 = m.addConstrs((ki_flag[i,t] == 0) >> (q0[t]+gp.quicksum(Q[t,s]*xi_hat[i,s]
                                                                    for s in range(model_params['T'])) <= 0.5)
                           for t in range(model_params['T'])
                          for i in range(model_params['N']))
    ki_cons2 = m.addConstrs((ki_flag[i,t] == 1) >> (q0[t]+gp.quicksum(Q[t,s]*xi_hat[i,s]
                                                                    for s in range(model_params['T'])) >= 0.500000000001)
                           for t in range(model_params['T'])
                          for i in range(model_params['N']))
    
    
    
    ### Objective function
    m.setObjective(

        # Weighted sum
        gp.quicksum(

            # i'th weight
            w_hat[i] * (

                gp.quicksum(model_params['k']*ki_flag[i,t] for t in range(model_params['T'])) + 
                
                
                # c^q
                gp.quicksum(model_params['c']*(q0[t]+gp.quicksum(Q[t,s]*xi_hat[i,s]
                                                 for s in range(model_params['T'])))
                            for t in range(model_params['T'])) + 

                # l^y_i
                gp.quicksum(model_params['l']*(yi0[i,t]+gp.quicksum(Yi[i,t,s]*xi_hat[i,s]
                                                    for s in range(model_params['T'])))
                            for t in range(model_params['T'])) 

            ) for i in range(model_params['N'])),        

        # min
        GRB.MINIMIZE
    )
    
    return m
```

```python
### Function to optimize model
def optimizeModel(m, xi_act, model_params):
   
    ### Optimize
    m.optimize()

    ### Optimal objective value
    v_opt = m.objVal

    ### Optimal ordering quantities
    q0_hat = [var.x
              for var in m.getVars()
              for t in range(model_params['T'])
              if (var.VarName == 'q0['+str(t)+']')]

    Q_hat = np.array([np.array([var.x 
                                for var in m.getVars()
                                for s in range(model_params['T'])
                                if (var.VarName == 'Q['+str(t)+','+str(s)+']')])
                      for t in range(model_params['T'])])

    q_hat = [q0_hat[t]+sum(Q_hat[t,s]*xi_act[s]
                           for s in range(model_params['T']))
             for t in range(model_params['T'])]

    return q_hat
```

```python
### Function to run rolling multi-policy static weighted SAA
def runMultiPolicyStaticWeightedSAA(Y_samples,
                                    weights_ij,
                                    sampleUpTo_start,
                                    params,
                                    model_type,
                                    decision_rules):

    # Get T horizon and R horizon rolling
    T_horizon = params['T_horizon']
    T_horizon_rolling = params['T_horizon_rolling']
    
    # Initialize starting inventory
    I_current=0
    
    # Initialize results lists and lists of ordering decisions
    I=list()
    q=list()
    I_q=list()
    y=list()
    I_q_y=list()
    q_hats=list()
    
    ## Iterate over full time horizon
    for t_current in range(T_horizon):
    
        # Set current model params
        model_params = dict({
            'T': min(T_horizon_rolling,T_horizon-t_current),
            'M': 2,
            'N': int(sampleUpTo_start+t_current),
            'c': params['c'],
            'k': params['k'],
            'l': 1,
            'h': params['h'],
            'b': params['b'],
            'decision_rules': decision_rules,
            'LogToConsole': params['LogToConsole'],
            'NonConvex': params['NonConvex'],
            'PSDTol': params['PSDTol'],
            'MIPGap': params['MIPGap']
        })
    
        
        # Samples of unknown variable (i,t)
        xi_hat = np.array(Y_samples.iloc[0:model_params['N'],0:model_params['T']])
        xi_act = np.array(Y_samples.iloc[model_params['N'],0:model_params['T']])

        # Sample weights (i)
        if model_type == 'WeightedSAA':
            w_hat = np.array(weights_ij[t_current])
        else:
            w_hat = np.array([np.repeat(1/model_params['N'],model_params['N'])]).flatten()
    
           
        # Setup model
        m = setUpModel(I_current, xi_hat, w_hat, model_params)

        # Optimize model
        q_hat = optimizeModel(m, xi_act, model_params)

        # Store results
        I = I+[I_current]
        q = q+[np.around(q_hat[0],0)]
        I_q = I_q+[I_current+np.around(q_hat[0],0)]
        y = y+[xi_act[0]]
        I_q_y = I_q_y+[I_current+np.around(q_hat[0],0)-xi_act[0]]
    
        # Update inventory
        I_current = I_current + np.around(q_hat[0],0) - xi_act[0]
        
        # Store q_hat
        q_hats = q_hats+[np.around(q_hat[0],0)]   
           

    ## Results
    results = pd.DataFrame({
        'SKU': np.repeat(params['SKU'],T_horizon),
        'T_horizon': np.repeat(T_horizon,T_horizon),
        'T_horizon_rolling': np.repeat(T_horizon_rolling,T_horizon),
        'T_horizon_k': np.repeat(T_horizon,T_horizon),
        'T_horizon_rolling_k': np.repeat(T_horizon_rolling,T_horizon),
        'SL': np.repeat(params['SL'],T_horizon),
        'IB': np.repeat(params['IB'],T_horizon),
        'k': np.repeat(params['k'],T_horizon),
        'c': np.repeat(params['c'],T_horizon),
        'h': np.repeat(params['h'],T_horizon),
        'b': np.repeat(params['b'],T_horizon),
        'j': np.array(range(1,T_horizon+1)),
        'I': I,
        'q': q,
        'I_q': I_q,
        'y': y,
        'I_q_y': I_q_y,
        'c_o': 0,
        'c_s': 0,
        'cost': 0
    })
    
    # Add cost
    results.c_o = (results.q>0) * results.k + results.q * results.c
    results.c_s = (results.I_q_y>0) * results.I_q_y * results.h + (results.I_q_y<0) * (-results.I_q_y) * results.b
    results.cost = results.c_o + results.c_s
    
    # Return
    return results, q_hats
```

```python
### Function to run rolling multi-policy ex-post model
def runMultiPolicyExPost(Y_samples,
                         sampleUpTo_start,
                         params):

    # Set current model params
    model_params = dict({
        'T': params['T_horizon'],
        'M': 2,
        'N': 1,
        'c': params['c'],
        'k': params['k'],
        'l': 1,
        'h': params['h'],
        'b': params['b'],
        'decision_rules': '',
        'LogToConsole': params['LogToConsole'],
        'NonConvex': params['NonConvex'],
        'PSDTol': params['PSDTol'],
        'MIPGap': params['MIPGap'],
        'IntFeasTol': params['IntFeasTol']
    })

    # Samples of unknown variable (i,t)
    xi_hat = np.array(Y_samples.iloc[sampleUpTo_start,0:model_params['T']])
    xi_act = np.array(Y_samples.iloc[sampleUpTo_start,0:model_params['T']])

    # Starting inventory
    I0=0

    ### Model coefficients

    # dim(A) = T x T x M; access via A[t,s,m]
    A = np.array([np.array([(np.array([model_params['h'],-model_params['b']])
                             if s==t
                             else np.array([model_params['h'],-model_params['b']]))
                            if s<=t
                            else np.array([0,0])
                            for s in range(model_params['T'])])
                  for t in range(model_params['T'])])

    # dim(B) = T x T x M; access via B[t,s,m]
    B = np.array([np.array([np.array([-model_params['h'],model_params['b']])
                            if s<=t
                            else np.array([0,0])
                            for s in range(model_params['T'])])
                  for t in range(model_params['T'])])

    # dim(C) = T x T x M; access via C[t,s,m]
    C = np.array([np.array([np.array([-1,-1])
                            if s==t
                            else np.array([0,0])
                            for s in range(model_params['T'])])
                  for t in range(model_params['T'])])

    # dim(f) = T x M; access via f[t,m]
    f = np.array([np.array([-model_params['h']*I0,model_params['b']*I0])
                  for t in range(model_params['T'])])


    
    ## Deinfe callback on optimality check and termination
    def cb(model, where):
        if where == GRB.Callback.MIPNODE:
            # Get model objective
            obj = model.cbGet(GRB.Callback.MIPNODE_OBJBST)

            # Has objective changed?
            if abs(obj - model._cur_obj) > 1e-8:
                # If so, update incumbent and time
                model._cur_obj = obj
                model._time = time.time()

        # Terminate if objective has not improved in 20s
        if time.time() - model._time > 20:
            model.terminate()


   
    ### Create new model
    m = gp.Model()
    m.setParam('LogToConsole', model_params['LogToConsole'])
    m.setParam('NonConvex', model_params['NonConvex'])
    m.setParam('PSDTol', model_params['PSDTol'])
    m.setParam('MIPGap', model_params['MIPGap']) 
    m.setParam('IntFeasTol', model_params['IntFeasTol']) 

    ## Primary decision variable (q)
    q0 = m.addVars(model_params['T'], lb=0, vtype='I', name='q0') # T vector

    ## Auxiliary decision variable (y x i)
    y0 = m.addVars(model_params['T'], vtype='I', name='y0') # T vector

    ## Fixed cost of ordering
    k_flag = m.addVars(model_params['T'], vtype='B') # T vector


    ### Add contraints -- non-negative decision space for primary decision rule
    q_cons = m.addConstrs(q0[t] >= 0
                          for t in range(model_params['T']))

    ### Add constraints -- model constraints
    cons = m.addConstrs(

        # A[t,.,m]*q0[t] where A[t,s,m]==0 for s > t
        gp.quicksum(A[t,s,m]*q0[s]
                    for s in range(model_params['T'])) + 

        # B[t,.,m]*xi[.] where B[t,s,m]==0 for s > t
        gp.quicksum(B[t,s,m]*xi_hat[s]
                    for s in range(model_params['T'])) + 

        # C[t,.,m]*y0[t] where C[t,s,m]==0 for s != t
        gp.quicksum(C[t,s,m]*y0[s]
                    for s in range(model_params['T']))

        <= f[t,m]

        for t in range(model_params['T'])
        for m in range(model_params['M'])
    )

    # Fixed ordering cost
    k_cons1 = m.addConstrs((k_flag[t] == 0) >> (q0[t] == 0)
                           for t in range(model_params['T']))
    k_cons2 = m.addConstrs((k_flag[t] == 1) >> (q0[t] >= 1)
                           for t in range(model_params['T']))



    ### Objective function
    m.setObjective(

        # k
        gp.quicksum(model_params['k']*k_flag[t] for t in range(model_params['T'])) + 


        # c^q
        gp.quicksum(model_params['c']*q0[t]
                    for t in range(model_params['T'])) + 

        # l^y
        gp.quicksum(model_params['l']*y0[t]
                    for t in range(model_params['T'])),       

        # min
        GRB.MINIMIZE
    )  

    # Last updated objective and time
    m._cur_obj = float('inf')
    m._time = time.time()

    m.optimize(callback=cb)
    
    ### Optimize
    #m.optimize()

    # Optimal objective value
    v_opt = m.objVal

    # Optimal ordering quantities
    q0_hats = [var.x
               for var in m.getVars()
               for t in range(model_params['T'])
               if (var.VarName == 'q0['+str(t)+']')]

    ### Store results
    
    # Initialize starting inventory
    I_current=0

    # Initialize results lists and lists of ordering decisions
    I=list()
    q=list()
    I_q=list()
    y=list()
    I_q_y=list()
    q_hats=list()

    # Iterate over results
    for t in range(model_params['T']):
        I = I+[I_current]
        q = q+[np.around(q0_hats[t],0)]
        I_q = I_q+[I_current+q0_hats[t]]
        y = y+[xi_act[t]]
        I_q_y = I_q_y+[I_current+np.around(q0_hats[t],0)-xi_act[t]]

        # Update inventory
        I_current = I_current + np.around(q0_hats[t],0) - xi_act[t]

        # Store q_hat
        q_hats = q_hats+[np.around(q0_hats[t],0)]   

    # Results
    results = pd.DataFrame({
        'SKU': np.repeat(params['SKU'],params['T_horizon']),
        'T_horizon': np.repeat(params['T_horizon'],params['T_horizon']),
        'T_horizon_rolling': np.repeat(params['T_horizon_rolling'],params['T_horizon']),
        'T_horizon_k': np.repeat(params['T_horizon'],params['T_horizon']),
        'T_horizon_rolling_k': np.repeat(params['T_horizon_rolling'],params['T_horizon']),
        'SL': np.repeat(params['SL'],params['T_horizon']),
        'IB': np.repeat(params['IB'],params['T_horizon']),
        'k': np.repeat(params['k'],params['T_horizon']),
        'c': np.repeat(params['c'],params['T_horizon']),
        'h': np.repeat(params['h'],params['T_horizon']),
        'b': np.repeat(params['b'],params['T_horizon']),
        'j': np.array(range(1,params['T_horizon']+1)),
        'I': I,
        'q': q,
        'I_q': I_q,
        'y': y,
        'I_q_y': I_q_y,
        'c_o': 0,
        'c_s': 0,
        'cost': 0
    })

    # Add cost
    results.c_o = (results.q>0) * results.k + results.q * results.c
    results.c_s = (results.I_q_y>0) * results.I_q_y * results.h + (results.I_q_y<0) * (-results.I_q_y) * results.b
    results.cost = results.c_o + results.c_s

    ### Return
    return results, q_hats
```

```python
# SKUs
SKU_range=range(1,460+1)

LogToConsole = 0
NonConvex = 2
PSDTol = 0
MIPGap = 0.0005
IntFeasTol = 0.0005
stopTime = 60



# For each SKU
for SKU in SKU_range:
    
    # Status
    print('SKU:', SKU)
    
    # Initialize results
    results = pd.DataFrame()

    # For each param setting
    for param_setting in range(len(param_settings)):
        
        # Status
        print('... param setting:', param_setting+1)
        
        # Get samples
        robj = pyreadr.read_r(PATH_SAMPLES+'/SKU'+str(SKU)+'/Static/TmpFiles'+str(1)+'/Y_samples_mv_k.RDS')
        Y_Samples = robj[None]

        # Get sampling
        robj = pyreadr.read_r(PATH_SAMPLES+'/SKU'+str(SKU)+'/Static/TmpFiles'+str(1)+'/sampleUpTo_start.RDS')
        sampleUpTo_start = robj[None].iloc[0,0]

        # Planning params
        params = dict({
            'SKU': SKU,
            'T_horizon': 13,
            'T_horizon_rolling': 13,
            'SL': param_settings['SL'].iloc[param_setting],
            'IB': param_settings['1-h/K'].iloc[param_setting],
            'k': param_settings['k'].iloc[param_setting],
            'c': param_settings['c'].iloc[param_setting],
            'h': param_settings['h'].iloc[param_setting],
            'b': param_settings['b'].iloc[param_setting],
            'LogToConsole': LogToConsole,
            'NonConvex': NonConvex,
            'PSDTol': PSDTol,
            'MIPGap': MIPGap,
            'IntFeasTol': IntFeasTol,
            'stopTime': stopTime
        })

        # Run
        result, q_hat = runMultiPolicyExPost(Y_Samples,
                                             sampleUpTo_start,
                                             params)

        results.append(result)  

        # Store
        if not results.empty:
            results = results.append(result)   
        else:
            results = pd.DataFrame(result) 
    
    results.to_csv(path_or_buf=PATH_RESULTS+'/SKU'+str(SKU)+'/ExPost_MultiPolicy.csv' , sep=',')

print('Done.')
```

```python
#### Update (merge) final results
PATH_RESULTS_OLD = PATH_RESULTS + '_OLD'
models = ['StaticSAA', 'StaticWeightedSAA', 'ExPost']

# SKUs
SKU_range=range(1,460+1)

# For each SKU
for SKU in SKU_range:
    
    # Status
    print('SKU:', SKU)
    
    for model in models:

        # Old results
        results_old = pd.read_csv(PATH_RESULTS_OLD+'/SKU'+str(SKU)+'/'+model+'_MultiPolicy.csv', sep=',', delimiter=',')
        results_old = results_old.iloc[:,1:results_old.shape[1]]
        
        # Attach old results to new results
        results_old.to_csv(path_or_buf=PATH_RESULTS+'/SKU'+str(SKU)+'/'+model+'_MultiPolicy.csv' , sep=',', mode='a', header=False)   
```

```python
# Initialize results
result = pd.DataFrame({})

# SKUs
SKU_range=range(1,460+1)

# Iterate over SKUs
for SKU in SKU_range:
    
    # Status
    print('SKU:', SKU)
    
    # Load and summarise model results
    res_SAA = pd.read_csv(PATH_RESULTS+'/SKU'+str(SKU)+'/StaticSAA_MultiPolicy.csv', sep=',', delimiter=',')
    res_SAA = res_SAA.groupby(['SKU', 'T_horizon_rolling', 'SL', 'IB', 'k']).agg(cost_SAA=('cost', sum)).reset_index()

    res_WSAA = pd.read_csv(PATH_RESULTS+'/SKU'+str(SKU)+'/StaticWeightedSAA_MultiPolicy.csv', sep=',', delimiter=',')
    res_WSAA = res_WSAA.groupby(['SKU', 'T_horizon_rolling', 'SL', 'IB', 'k']).agg(cost_WSAA=('cost', sum)).reset_index()

    res_ExPost = pd.read_csv(PATH_RESULTS+'/SKU'+str(SKU)+'/ExPost_MultiPolicy.csv', sep=',', delimiter=',')
    res_ExPost = res_ExPost.groupby(['SKU', 'SL', 'IB', 'k']).agg(cost_ExPost=('cost', sum)).reset_index()  

    # Merge results per model
    res = pd.merge(pd.merge(res_SAA,
                        res_WSAA,
                        on=['SKU', 'T_horizon_rolling', 'SL', 'IB', 'k'],
                        how='left'),
               res_ExPost,
               on=['SKU', 'SL', 'IB', 'k'],
               how='left')

    # Attach to full data frame
    result = pd.concat([result,res])
```

```python
## Get coefficient of prescriptiveness
result['Pq'] = 1 - (result.cost_WSAA - result.cost_ExPost) / (result.cost_SAA - result.cost_ExPost)
result.loc[result.cost_WSAA == result.cost_SAA, 'Pq'] = 0   
```

```python
result.loc[result.k==100].groupby(['T_horizon_rolling', 'SL', 'IB', 'k']).agg(Pq = ('Pq', np.median),
                                                                              Pq_share = ('Pq', lambda x: sum(x>0)/len(x)))
```

```python
result_summary = result.loc[result.k==100].groupby(['SKU', 'SL', 'IB', 'k']).agg(cost_SAA = ('cost_SAA', min),
                                                                                 cost_WSAA = ('cost_WSAA', min),
                                                                                 cost_ExPost = ('cost_ExPost', min)).reset_index()
result_summary
```

```python
## Get coefficient of prescriptiveness
result_summary['Pq'] = 1 - (result_summary.cost_WSAA - result_summary.cost_ExPost) / (result_summary.cost_SAA - result_summary.cost_ExPost)
result_summary.loc[result_summary.cost_WSAA == result_summary.cost_SAA, 'Pq'] = 0   

Pq_x_SL = result_summary.loc[result_summary.k==100].groupby(['SL']).agg(Pq_median = ('Pq', np.median),
                                                                        Pq_mean = ('Pq', np.mean),
                                                                        Pq_share = ('Pq', lambda x: sum(x>0)/len(x))).reset_index()
Pq_x_SL
```
